# Reference date = next day after last order

```python
data.head()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
reference_date = data['Order Date'].max() + pd.Timedelta(days=1)
```

```python
reference_date
```

```python
recency = data.groupby('Customer ID')['Order Date'].max().reset_index()
```

```python
recency['Recency'] = (reference_date - recency['Order Date']).dt.days
```

```python
frequency = data.groupby('Customer ID')['Order ID'].nunique().reset_index()
frequency.rename(columns={'Order ID': 'Frequency'}, inplace=True)
```

```python
monetary = data.groupby('Customer ID')['Sales'].sum().reset_index()
monetary.rename(columns={'Sales': 'Monetary'}, inplace=True)
```

```python
rfm = recency.merge(frequency, on='Customer ID').merge(monetary, on='Customer ID')
```

# Keep only the score columns

```python
rfm = rfm[['Customer ID', 'Recency', 'Frequency', 'Monetary']]
rfm.head()
```

Recency Score (lower recency = better)

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
```

Frequency Score (higher frequency = better)

```python
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
```

Monetary Score (higher spending = better)

```python
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

Combine to create an RFM score

```python
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm.head()
```

```python
def rfm_segment(row):
    r, f, m = int(row['R_Score']), int(row['F_Score']), int(row['M_Score'])

    if r >= 4 and f >= 4 and m >= 4:
        return "High-Value Customer"
    elif r >= 4 and f >= 3:
        return "Potential Loyalist"
    elif r <= 2 and f >= 3:
        return "At-Risk Customer"
    elif r == 1:
        return "Lost Customer"
    else:
        return "Regular Customer"

rfm['Segment'] = rfm.apply(rfm_segment, axis=1)
rfm.head()
```

#Assign Customers to each Segment

```python
def assign_segment(row):
    r = int(row['R_Score'])
    f = int(row['F_Score'])
    m = int(row['M_Score'])

    # High-Value Customers
    if r == 5 and f == 5 and m == 5:
        return "High-Value Customer"

    # Loyal Customers
    if (f >= 4 and m >= 4 and r >= 3):
        return "Loyal Customer"

    # Potential Loyalists
    if (f >= 3 and m >= 3 and r <= 3):
        return "Potential Loyalist"

    # At-Risk Customers
    if (r <= 2 and f >= 3):
        return "At-Risk Customer"

    # New Customers
    if (r == 5 and f <= 2 and m <= 2):
        return "New Customer"

    # Lost Customers
    if (r == 1 and f == 1 and m == 1):
        return "Lost Customer"

    # Default
    return "Regular Customer"

# Apply the function
rfm['Segment'] = rfm.apply(assign_segment, axis=1)

rfm.head()
```
